# 🌱 Wanderland

Write simple Python commands and watch your character animate through a stylized 3D world.
Collect all gems and land on the goal ring to win!

```
# Create and show the world
# It has its own ▶ Run My Code button right inside the 3D scene
import wanderland as mp
from wanderland import move_forward, turn_left, turn_right, collect_gem

# We'll use the built-in "gem_path" puzzle
world = mp.World(mp.puzzles.gem_path())
world
```

## Your program

The character starts in the top-left corner. Two gems wait at opposite corners, with a pond in the middle you must walk around.

Walk the character *onto* a gem's tile, then call `collect_gem()` to pick it up.

The commands this world allows (its *action space*):
- `move_forward()`
- `turn_left()` / `turn_right()`
- `collect_gem()`

Define the steps in `solution()`. Then run this cell to load it. Finally, press **▶ Run My Code** in the 3D scene above!

```
def solution():
    move_forward()
    move_forward()
    move_forward()
    collect_gem()   # first gem

    turn_right()
    move_forward()
    move_forward()

    turn_right()
    move_forward()
    move_forward()
    move_forward()
    collect_gem()   # second gem

    turn_right()
    turn_right()

    move_forward()
    move_forward()
    # move_forward() # Uncomment this to reach the goal!

world.load(solution)
```

### Check your results
Run the cell below after the animation finishes to check your score.

```
print(f"Gems collected: {world.gems_collected} / {world.total_gems}")
print(f"Reached goal: {world.reached_goal}")
if world.success:
    print("🎉 Puzzle solved!")
else:
    print("Keep trying...")
```
